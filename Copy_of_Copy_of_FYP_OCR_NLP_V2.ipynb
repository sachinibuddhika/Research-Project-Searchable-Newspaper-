```python
# Import required modules
from paddleocr import PaddleOCR, draw_ocr
import os
import requests
import json
import mysql.connector

```

```python
# Initialize the OCR model
ocr = PaddleOCR(det_algorithm='DB', rec_algorithm='CRNN', use_angle_cls=True, lang='en')
```

```python
# Verify the dataset path
if not os.path.exists(data_set_path):
    print(f"Directory not found: {data_set_path}")
else:
    # List to store OCR results
    ocr_results = []

    # Loop through all images in the directory
    for image_name in os.listdir(data_set_path):
        image_path = os.path.join(data_set_path, image_name)

        if os.path.isfile(image_path):
            # Process the image and extract text
            text_result = ocr.ocr(image_path, cls=True)

            # Collect the extracted text for the current image
            image_texts = [line[1][0] for line in text_result[0]]

            # Combine the text lines into a single string
            text_to_correct = '\n'.join(image_texts)

            # Add the OCR output to the list
            ocr_results.append((image_name, text_to_correct))

    # Print all OCR results after processing all images
    for image_name, text_to_correct in ocr_results:
        print(f"Text in {image_name}:")
        print(text_to_correct)
        print("\n" + "-"*50 + "\n")
```

```python
# Function to send request to OpenAI API (ChatGPT)
def call_openAI(prompt, openai_api_key):
    URL = "https://api.openai.com/v1/chat/completions"  #endpoint for ChatGPT

    payload = {
        "model": "gpt-4-turbo",
        "messages": [{"role": "user", "content": prompt}],
        "temperature": 0.0,
        "top_p": 1.0,
        "n": 1,
        "stream": False,
        "presence_penalty": 0,
        "frequency_penalty": 0,
    }

    headers = {
        "Content-Type": "application/json",
        "Authorization": f"Bearer {openai_api_key}"
    }

    # Send request to OpenAI API to correct the OCR text
    response = requests.post(URL, headers=headers, json=payload)

    if response.status_code == 200:
        data = response.json()
        corrected_text = data['choices'][0]['message']['content'].strip()
        return corrected_text
    else:
        print(f"Error: {response.status_code}")
        print(response.text)
        return None
```
